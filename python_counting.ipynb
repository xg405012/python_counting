```python
fastq= f.readlines()
print (fastq)
```

```python
count_list = [0,0,0,0,]
for line_num, line in enumerate(fastq):#go through each line
    if line_num % 4 ==1:
            #check if it is a sequence
        for base in line:#go through each base so we can count it
            if base == 'A':
                #incrementally increase the A item in the list
                count_list[0]=count_list[0]+1
            if base == 'C':
                #incrementally increase the A item in the list
                count_list[1]=count_list[1]+1
            if base == 'G':
                 count_list[2]=count_list[2]+1
            if base == 'T':
                 count_list[3]=count_list[3]+1
                
print (count_list)
```

```python
def count_kmerknown(filename):
    """
    list all pair
    starting from line and line number
    then check if it is a sequence,cause we need to go through each line
    the reason I add x here is that after I ran without-x model I found errors, so I figured out if it goes to the last one,
    there is no number to +1, so I just gave an chracter to the end of the base
    a+b make sure the results are pairs not sing number
    then print
    """
    count_base={'AA':0, 'CC':0, 'GG':0, 'TT':0, 'AC':0, 'AG':0, 'AT':0, 'CG':0, 'CT':0, 'GT':0, 'CA':0, 'GA':0, 'TA':0, 'GC':0, 'TC':0, 'TG':0}
    for line_num, line in enumerate(filename):#go through each line
        if line_num % 4 ==1:#check if it is a sequence
            l=line+'x'
            for base_num, base in enumerate(line):
                a=l[base_num]
                b=l[(base_num)+1]
                base_=str(a+b)#got the pairs not the single number
                if base_ == 'AA':
                    count_base['AA'] +=1
                elif base_ == 'CC':
                    count_base['CC'] +=1
                elif base_ == 'GG':
                    count_base['GG'] +=1
                elif base_ == 'TT':
                    count_base['TT'] +=1
                elif base_ == 'AC':
                    count_base['AC'] +=1
                elif base_ == 'AG':
                    count_base['AG'] +=1
                elif base_ == 'AT':
                    count_base['AT'] +=1
                elif base_ == 'CG':
                    count_base['CG'] +=1
                elif base_ == 'CT':
                    count_base['CT'] +=1
                elif base_ == 'GT':
                    count_base['GT'] +=1
                elif base_ == 'CA':
                    count_base['CA'] +=1
                elif base_ == 'GA':
                    count_base['GA'] +=1
                elif base_ == 'TA':
                    count_base['TA'] +=1
                elif base_ == 'GC':
                    count_base['GC'] +=1
                elif base_ == 'TC':
                    count_base['TC'] +=1
                elif base_ == 'TG':
                    count_base['TG'] +=1              
    print (count_base)
count_kmerknown(fastq)
```

```python
def count_unknownkmers(filename):
    """
    cause I don't know what pairs are in the sequence at this time, I am going to check from the start, if it is in the dictionary
    the dictionary will +1, then I will got all the pairs
    """
    counterr ={} # a empty dictionary
    for line_num, line in enumerate(filename):#go through each line   
        line=line.rstrip() #this just easily solve the problem I countered in the above question    
        if line_num % 4 ==1:#check if it is a sequence
            for base_num, base in enumerate(line):
                base_=line[base_num:base_num+2]
                if base_ in counterr:
                    counterr[base_]+=1
                else:
                    counterr[base_]=1
    print(counterr)
count_unknownkmers(fastq)
```

```python
def count_kmers(filename,n):
    """
    if i want the fuction fits all sizes, I just need to change the fixed number to 'n', but I need to state which size I 
    want after the filename, this time I choose 7
    """
    counterr ={} #a empty dictionary
    for line_num, line in enumerate(filename):#go through each line   
        line=line.rstrip()     
        if line_num % 4 ==1:#check if it is a sequence
            for base_num, base in enumerate(line):
                base_=line[base_num:base_num+n]
                if base_ in counterr:
                    counterr[base_]+=1
                else:
                    counterr[base_]=1
    print(counterr)
count_kmers(fastq,7)
```
